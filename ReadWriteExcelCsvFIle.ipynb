# Read csv file 

```python
import pandas as pd  #In these exaple we use stock_data.csv file
df = pd.read_csv('stock_data.csv')
df
```

```python
import pandas as pd #In this ex we see its treat first row as header but we want to skip first row.
df = pd.read_csv('stock_data.csv')
df
```

```python
import pandas as pd #In this example you see it skipped first row of data
df = pd.read_csv('stock_data.csv', skiprows=1)
df
```

```python
import pandas as pd #In this example you see it skipped first and second row and more rows of data
df = pd.read_csv('stock_data.csv', skiprows=2)
df
```

```python
import pandas as pd #In this example you see it make the row as header row 
df = pd.read_csv('stock_data.csv', header=1)
df
```

```python
import pandas as pd #In this example you see when you dont want any header row 
df = pd.read_csv('stock_data.csv', header=None)
df
```

```python
import pandas as pd #In this example we give column name when data dont have header row
df = pd.read_csv('stock_data.csv', header=None, names=["tickers","eps","revenue","price","people"])
df
```

```python
import pandas as pd #to see specific number of rows
df = pd.read_csv('stock_data.csv', nrows=3)
df
```

```python
import pandas as pd #when you not available or n.a. replace with NAN and this is helpful in cleaning of data
df = pd.read_csv('stock_data.csv', na_values=["not available","n.a."])
df
```

```python
#when you not available or n.a. replace with NAN and this is helpful in cleaning of data
#to change in  multiple column

import pandas as pd 
df = pd.read_csv('stock_data.csv', na_values={
    "eps":["not available","n.a."],
    "revenue":["not available","n.a.",-1]
})
df
```

# Write data in a csv file

```python
#it will create a new csv  file with data of df
df.to_csv('new.csv')
```

# Read data in Excel

```python
import pandas as pd
df=pd.read_excel('stock_data.xlsx', "Sheet1")
df
```

```python

import pandas as pd
def converts_people_cell(cell):
    if cell=="n.a.":
        return "rajat"
    return cell
df=pd.read_excel('stock_data.xlsx', "Sheet1", converters={
    "people": converts_people_cell
})
df
```

# Write data in excel


```python
df.to_excel("new.xlsx",sheet_name="stocks")
```
